Importing the Dependencies 

```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
#loading the dataset to Pandas DataFrame 
credit_card_data = pd.read_csv ('/creditcard.csv')
credit_card_data
```

```python
credit_card_data.info()
```

```python
#Checking the number of missing values in each columns 
credit_card_data.isnull().sum()
```

```python
# Distribution of legit and fraudulent transactions in the dataset
credit_card_data['Class'].value_counts()
```

This is a highly unbalanced dataset with 284315 normal /legit transactions and 492 fraudulent transactions.

```python
#Seperating the data for analysis 
legit = credit_card_data [credit_card_data.Class == 0]
fraud = credit_card_data [credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare teh values for both transactions 
credit_card_data.groupby("Class").mean()
```

Under-sampling the data : building a sample data set that consists of similar distribution of legit and fraudulent transactions.

Here, number of fraudulent transactions = 492


```python
legit_sample = legit.sample(n = 492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the datase into Features and Targets 

```python
X = new_dataset.drop(columns = 'Class', axis =1)
Y = new_dataset ['Class']
```

```python
print(Y)
```

Split the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

Model Training 

Logistic Reregression 


```python
# Model Training 
model = LogisticRegression()
```

```python
# training the logistic regression model with training data 
model.fit(X_train, Y_train)
```

Model evaluation through accuracy score

```python
# Accuracy on the training data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy of the Training Data : ', training_data_accuracy)
```

```python
#accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the Test Data : ', test_data_accuracy)
```
